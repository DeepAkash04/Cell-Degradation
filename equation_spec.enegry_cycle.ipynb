```python
import numpy as np
import matplotlib.pyplot as plt
import pybamm as py
import pandas as pd 
from datetime import timedelta
import seaborn as sns
%matplotlib inline

df=pd.read_excel("C:\\Users\\Harisankar.Suresh\\Documents\\AKASHDEEP\\A1C2C0088854.xlsx",sheet_name='step')
df2=df[df['Step Index']==2]
df3 = df2.iloc[20:32]
df3
df3.columns
x=df3['Cycle Index']
y=df3['Chg. Spec. Energy(mWh/g)']
plt.plot(x,y)




import numpy as np

```

```python
df4 = pd.DataFrame({'x': x, 'y': y})
window_size=3
df4['y_sma'] = df4['y'].rolling(window_size).mean()
df4.dropna(inplace=True)
y_sma=df4['y_sma']
x1=df4['x']
plt.plot(df4.x, df4.y, color = 'red', label = 'Original')
plt.plot(df4.x,df4.y_sma, color = 'green', label = 'Moving Average')
plt.legend()

plt.plot(x1,y_sma,label="moving_avg")
plt.plot(x,y, label = "line 2")
```

```python
# Perform polynomial regression of degree n
n = 10
coeff = np.polyfit(x, y, n)
coeff
 #Print the equation
poly = np.poly1d(coeff)
print(poly)
y_pre=[]
for i in df4.x:
    
    y_pre.append(poly(i))

y_pre
plt.plot(df4.x, df4.y, color = 'red', label = 'Original')
#plt.plot(df4.x,df4.y_sma, color = 'green', label = 'Moving Average')
plt.legend()
plt.plot(df4.x,y_pre)
```
